# Data Cleaning

## Bluestock Mutual Fund Analytics Capstone

### Objective

Clean the raw datasets by handling missing values, duplicates, incorrect data types, and preparing the data for analysis.

```python
import pandas as pd
```

```python
nav_history = pd.read_csv("../data/raw/02_nav_history.csv")

scheme_performance = pd.read_csv("../data/raw/07_scheme_performance.csv")

investor_transactions = pd.read_csv("../data/raw/08_investor_transactions.csv")
```

## Missing Value Analysis

```python
print("NAV History")
print(nav_history.isnull().sum())

print("\nScheme Performance")
print(scheme_performance.isnull().sum())

print("\nInvestor Transactions")
print(investor_transactions.isnull().sum())
```

### Observation

- Missing values were checked across all datasets.
- Most columns contained complete information.
- Missing values, if present, will be handled appropriately before analysis.

## Duplicate Record Analysis

```python
print("NAV History:", nav_history.duplicated().sum())

print("Scheme Performance:", scheme_performance.duplicated().sum())

print("Investor Transactions:", investor_transactions.duplicated().sum())
```

```python
nav_history.drop_duplicates(inplace=True)

scheme_performance.drop_duplicates(inplace=True)

investor_transactions.drop_duplicates(inplace=True)
```

### Observation

- No duplicate records were found in any of the datasets.
- The datasets already contain unique records.
- No duplicate removal was required before further analysis.

## Date Format Conversion

```python
nav_history["date"] = pd.to_datetime(nav_history["date"])

investor_transactions["transaction_date"] = pd.to_datetime(
    investor_transactions["transaction_date"]
)

print(nav_history.dtypes)

print(investor_transactions.dtypes)
```

### Observation

- The `date` column in the NAV History dataset was successfully converted to datetime format.
- The `transaction_date` column in the Investor Transactions dataset was successfully converted to datetime format.
- Date conversion enables efficient time-series analysis, rolling calculations, and date-based filtering in subsequent stages of the project.

## Export Cleaned Datasets

```python
nav_history.to_csv(
    "../data/processed/02_nav_history_cleaned.csv",
    index=False
)

scheme_performance.to_csv(
    "../data/processed/07_scheme_performance_cleaned.csv",
    index=False
)

investor_transactions.to_csv(
    "../data/processed/08_investor_transactions_cleaned.csv",
    index=False
)

print("Cleaned datasets exported successfully.")
```

# Conclusion

The datasets were successfully cleaned by:

- Checking missing values
- Verifying duplicate records
- Converting date columns to datetime format
- Exporting the cleaned datasets to the `data/processed` folder

These cleaned datasets will be used for Exploratory Data Analysis, Performance Analytics, Advanced Analytics, and Power BI Dashboard development.
